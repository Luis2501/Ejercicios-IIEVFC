Considerar un aro horizontal sin rozamiento de radio R. En este aro enhebro tres objetos con masas idénticas m, y, entre los objetos, tres muelles idénticos, cada uno con una masa constante de fuerza k.

Sabemos que, la energía cinética es:

$$T=\frac{1}{2}mR^2(\dot{\phi_1^2}+\dot{\phi_2^2}+\dot{\phi_3^2})$$

y la energía potencial:

$$U=\frac{1}{2}kR^2 [(\phi_1-\phi_2)^2+(\phi_2-\phi_3)^2+(\phi_3-\phi_1)^2]$$

Resultado: Sistema de EDOs 

$$m\frac{d^2\phi_1}{dt^2}=-k(\phi_1-\phi_2)-k(\phi_1-\phi_3)$$

$$m\frac{d^2\phi_2}{dt^2}=-k(\phi_2-\phi_3)-k(\phi_2-\phi_1)$$

$$m\frac{d^2\phi_3}{dt^2}=-k(\phi_3-\phi_1)-k(\phi_3-\phi_2)$$

a) Encontrar las ecuaciones de movimiento para cada masa usando las ecuaciones de Euler-Lagrange. *(ejemplo en PCP_sesión5.ipynb)*

b) Resolver numéricamente el sistema de ecuaciones con los métodos de Euler y RK de orden 4. Para los datos: $\phi_1(t=0)=\phi_2/2,\quad \phi_2(t=0)=\phi_3/2,\quad \phi_3(t=0)=\pi/10 \quad $ y $\quad \phi_1'(t=0)=1,\quad \phi_2'(t=0)=1,\quad \phi_3'(t=0)=1, \quad R=m=k=1$


c) Mostrar los resultados y explicar la dinámica del sistema en base al resultado de la simulación.

## Solución 

____________

El lagrangiano del sistema es

$$L = T - U =\frac{1}{2}mR^2(\dot{\phi_1^2}+\dot{\phi_2^2}+\dot{\phi_3^2}) - \frac{1}{2}kR^2 [(\phi_1-\phi_2)^2+(\phi_2-\phi_3)^2+(\phi_3-\phi_1)^2]$$

Y mediante las ecuaciones de Euler-Lagrange, 

$$\frac{d}{dt} \left( \frac{\partial L}{ \partial \dot{ \phi_{i}} } \right) = \frac{\partial L}{ \partial  \phi_{i} } \qquad \qquad  i = 1,2,3$$

Por la forma que tiene el lagrangiano, tenemos que

$$\frac{d}{dt} \left( \frac{\partial T}{ \partial \dot{ \phi_{i}} } \right) = - \frac{\partial U}{ \partial  \phi_{i} } \qquad \qquad  i = 1,2,3$$

No es muy dificil observar que 

$$\frac{d}{dt} \left( \frac{\partial T}{ \partial \dot{ \phi_{i}} } \right) = mR^{2} \ddot{\phi_{i}} \qquad \qquad  i = 1,2,3$$

En cambio para el lado derecho de la ecuación de Euler-Lagrange, haremos uno por uno, entonces

$$\frac{\partial U}{\partial \phi_1} = kR^2[(\phi_1-\phi_2)+ k(\phi_1-\phi_3)]$$

$$\frac{\partial U}{\partial \phi_2} = kR^2[(\phi_2-\phi_3) +(\phi_2-\phi_1)]$$

$$\frac{\partial U}{\partial \phi_3} = kR^2[(\phi_3-\phi_1)+(\phi_3-\phi_2)]$$

Por lo tanto, 

$$ mR^{2} \ddot{\phi_{1}} = -kR^2[(\phi_1-\phi_2)+ k(\phi_1-\phi_3]$$

$$ mR^{2} \ddot{\phi_{2}} = -kR^2[(\phi_2-\phi_3) +(\phi_2-\phi_1)]$$

$$ mR^{2} \ddot{\phi_{3}} = -kR^2[(\phi_3-\phi_1)+(\phi_3-\phi_2)]$$

Que podemos reescribir como

$$m\frac{d^2\phi_1}{dt^2}=-k(\phi_1-\phi_2)-k(\phi_1-\phi_3)$$

$$m\frac{d^2\phi_2}{dt^2}=-k(\phi_2-\phi_3)-k(\phi_2-\phi_1)$$

$$m\frac{d^2\phi_3}{dt^2}=-k(\phi_3-\phi_1)-k(\phi_3-\phi_2)$$

que es lo que queriamos demostrar.

Para resolver de manera númerica, primero importamos las librerias necesarias a utilizar. 

```python
import plotly.graph_objects as go
import numpy as np
import sys

sys.path.append("../")

from PhysicsPy.ODEsMethods import *
```

Se importa el módulo `ODEsMethods` de `PhysicsPy`, ahi se encuentran los métodos a utilizar.

Ahora, crearemos una clase `Aro` que nos permita crear `aross` con distintas características, en particular `k` y `m`.

```python
class Aro:

    def __init__(self, k, m):

        self.k, self. m = k, m 

    def __call__(self, u, t):

        Phi_1, Vphi_1, Phi_2, Vphi_2, Phi_3, Vphi_3 = u
        k, m = self.k, self.m
        
        return np.array([Vphi_1, (k/m)*(-(Phi_1 - Phi_2) - (Phi_1 - Phi_3)),
                        Vphi_2, (k/m)*(-(Phi_2 - Phi_3) - (Phi_2 - Phi_1)),
                        Vphi_3, (k/m)*(-(Phi_3 - Phi_1) - (Phi_3 - Phi_2)) ])
```

Creamos el `aro` con las condiciones que nos imponen.

Considerando las condiciones inciales y mediante el módulo `ODEsMethods` iteraremos en cada método para obtener las distintas soluciones.

```python
aro = Aro(k = 1, m = 1)

Conditions = [np.pi/40, 1, np.pi/20, 1, np.pi/10, 1]
Solutions = []

for solver_class in [Euler, Runge_Kutta]:

    Solucion = Runge_Kutta(aro)
    Solucion.InitialConditions(Conditions, [0, 5], 1e-3)
    Phi, t = Solucion.SolveODE()
    
    Solutions.append(Phi)

Solutions = np.array(Solutions)
```

Dadas las soluciones, nos queda generar las gráficas, para eso crearemos las figuras para después mostrarlas.

```python
Figuras = []

for k, Method in zip(range(2), ["Euler", "Runge Kutta"]):

    fig = go.Figure()

    for i in range(0,6,2):

        fig.add_trace(go.Scatter(x =t, y= Solutions[k][:,i], name = f"$\phi_{i + 1}$"))

    fig.update_layout(title="Movimiento del sistema (Método de " + Method +")", 
                      xaxis_title="tiempo (s)", yaxis_title = r"$\phi$")

    Figuras.append(fig)
    
    del fig
```
